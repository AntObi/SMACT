## Compound counting
### Search through possible binary, ternary and quaternary combinations
### Various levels of screening
#### Note : in the example we do only the first 10 elements, full periodic table search takes a long time

```python
%load_ext autoreload
%autoreload 2
import smact.core as core
import matplotlib.pyplot as plt
import numpy as np
```

### Set up a list using Josh's ordered_elements function

```python
list_of_elements = core.ordered_elements(1,10)
```

### Set up the counters
* element_count : the count if only elements (not ions) are considered
* ion_count : the count if the various ions of each element is considered
* charge_neutral_count : filter out the combinations of ions which are not charge neutral
* pauling_count : the count when we apply Andy's Pauling sense filter

```python
count_of_ternary = 0
count_raw_ternary = 0
element_count = 0
ion_count = 0
charge_neutral_count = 0
pauling_count = 0
```

### Loop through the list for binaries

```python
for i, ele_a in enumerate(list_of_elements):
        element_count = element_count + 1
        paul_a = core.Element(ele_a).pauling_eneg
        for ox_a in core.Element(ele_a).oxidation_states:
                ion_count = ion_count + 1
                for j, ele_b in enumerate(list_of_elements[i+1:]):
                        element_count = element_count + 1
                        paul_b = core.Element(ele_b).pauling_eneg
                        for ox_b in core.Element(ele_b).oxidation_states:
                                ion_count = ion_count + 1
                                element = [ele_a, ele_b]
                                oxidation_states = [ox_a, ox_b]
                                pauling_electro = [paul_a, paul_b]
                                electroneg_makes_sense = core.pauling_test(oxidation_states, pauling_electro)
                                cn_e, cn_r = core.charge_neutrality([ox_a, ox_b],threshold=1)
                                if cn_e:
                                        charge_neutral_count = charge_neutral_count + 1
                                        if electroneg_makes_sense:
                                                pauling_count = pauling_count + 1
```

```python
print 'Element count: ', element_count
print 'Ion count: ', ion_count
print 'Charge neutral: ', charge_neutral_count
print 'Pauling sensible: ', pauling_count
```
